```python
import pandas as pd

df = pd.read_excel("Dataset for Data Analytics (1).xlsx")
df.head()
```

```python
from pandasql import sqldf

pysqldf = lambda q: sqldf(q, globals())
```

```python
pysqldf("SELECT * FROM df LIMIT 5")
```

```python
pysqldf("""
SELECT Product, COUNT(*) as count
FROM df
GROUP BY Product
""")
```

```python
pysqldf("""
SELECT Product, SUM(TotalPrice) as revenue
FROM df
GROUP BY Product
ORDER BY revenue DESC
""")
```

```python
pysqldf("""
SELECT PaymentMethod, COUNT(*) as total
FROM df
GROUP BY PaymentMethod
""")
```

```python
pysqldf("""
SELECT OrderStatus, COUNT(*) as total
FROM df
GROUP BY OrderStatus
""")
```

```python
pysqldf("""
SELECT * FROM df
WHERE Product = 'Laptop'
""")
```
